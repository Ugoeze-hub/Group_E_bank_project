```python

user_data = {
    "13579": 2000,
    "12345": 3000,
    "24680": 3000000,
    "98765": 40
}

def main():
    print("Welcome to Spark Financial Organization")
    print("Please, how can I help you?")
    print("""
       _____
       |#$|||
       @!#$%|
    """)
    print("Please provide the needed information.")

    max_attempts = 5
    attempts = 0

    while attempts < max_attempts:
        u_pin = input("Enter your PIN: ")
        if u_pin in user_data:  
            print("Correct. You are logged into our Spark portal.")
            main_menu(u_pin)  
            break
        else:
            attempts += 1
            remaining_attempts = max_attempts - attempts
            if remaining_attempts == 0:
                print("Account blocked. Please visit X_coders.")
            else:
                print(f"Invalid PIN. {remaining_attempts} trial(s) left.")

def main_menu(user_pin):
    while True:
        print("Main Menu:")
        print("1. Financial Services")
        print("2. Customer Care")
        print("3. Other Services")
        print("4. Logout")

        try:
            choice = int(input("Please select an option (1, 2, 3, or 4): "))
            if choice == 1:
                financial_services(user_pin)
            elif choice == 2:
                print("Customer care selected. How can we assist you?")
            elif choice == 3:
                print("Other services selected.")
            elif choice == 4:
                print("You have logged out. Thank you for using Spark Financial Organization!")
                break
            else:
                print("Invalid selection. Please choose a valid option.")
        except ValueError:
            print("Invalid input! Please enter a number (1, 2, 3, or 4).")

def financial_services(user_pin):
    while True:
        print("\nFinancial Services:")
        print("1. Transfer")
        print("2. Balance")
        print("3. Account Details")
        print("4. Back to Main Menu")

        try:
            fs_choice = int(input("Please select an option (1, 2, 3, or 4): "))
            if fs_choice == 1:
                transfer(user_pin)
            elif fs_choice == 2:
                print(f"Your balance is: {user_data[user_pin]} BITCOIN")
            elif fs_choice == 3:
                print("Account details selected. Retrieving details...")
                print("Account Name: John Doe")
                print("Account Number: 1234567890")
            elif fs_choice == 4:
                break 
            else:
                print("Invalid selection. Please choose 1, 2, 3, or 4.")
        except ValueError:
            print("Invalid input! Please enter a valid number.")

def transfer(user_pin):
    while True:
        try:
            amount = int(input("Enter the amount you want to transfer: "))
            if amount > user_data[user_pin]:
                print("Insufficient balance! Transfer cannot be processed.")
            else:
                account_no = input("Enter the account number: ")
                if not account_no.isdigit():
                    print("Invalid account number. It must be numeric.")
                    continue

                bank = input("Enter the bank name: ")
                if bank.isdigit():
                    print("Bank name cannot be numeric. Please enter a valid name.")
                    continue

                name = input("Enter the recipient's name: ")
                if name.isdigit():
                    print("Recipient name cannot be numeric. Please enter a valid name.")
                    continue

                user_data[user_pin] -= amount
                print(f"Transfer of {amount} BITCOIN to {name} at {bank} is being processed.")
                print(f"Transfer successful! Your new balance is: {user_data[user_pin]} BITCOIN")
                break
        except ValueError:
            print("Invalid input! Please enter valid details.")


main()
```

```python
import tkinter as tk
from tkinter import messagebox

# User data with PINs and balances
user_data = {
    "13579": 2000,
    "12345": 3000,
    "24680": 3000000,
    "98765": 40
}

# Global variable to track logged-in user
logged_in_pin = None


# Function to handle login
def login():
    global logged_in_pin
    pin = pin_entry.get()

    if pin in user_data:
        logged_in_pin = pin
        messagebox.showinfo("Login Successful", "You are logged into Spark Portal.")
        login_window.destroy()
        show_main_menu()
    else:
        attempts.set(attempts.get() - 1)
        if attempts.get() == 0:
            messagebox.showerror("Account Blocked", "Account blocked. Please visit X_Coders.")
            login_window.destroy()
        else:
            messagebox.showerror("Invalid PIN", f"Invalid PIN. {attempts.get()} trial(s) left.")


# Function to show the main menu
def show_main_menu():
    main_menu = tk.Tk()
    main_menu.title("Main Menu")

    tk.Label(main_menu, text="Main Menu", font=("Helvetica", 16, "bold")).pack(pady=10)

    tk.Button(main_menu, text="Financial Services", width=20, command=lambda: financial_services(main_menu)).pack(pady=5)
    tk.Button(main_menu, text="Customer Care", width=20, command=lambda: messagebox.showinfo("Customer Care", "How can we assist you?")).pack(pady=5)
    tk.Button(main_menu, text="Other Services", width=20, command=lambda: messagebox.showinfo("Other Services", "No other services available.")).pack(pady=5)
    tk.Button(main_menu, text="Logout", width=20, command=main_menu.destroy).pack(pady=10)

    main_menu.mainloop()


# Function to handle financial services menu
def financial_services(parent_window):
    parent_window.destroy()

    services_window = tk.Tk()
    services_window.title("Financial Services")

    tk.Label(services_window, text="Financial Services", font=("Helvetica", 16, "bold")).pack(pady=10)

    tk.Button(services_window, text="Transfer", width=20, command=lambda: transfer(services_window)).pack(pady=5)
    tk.Button(services_window, text="Balance", width=20, command=lambda: messagebox.showinfo("Balance", f"Your balance is: {user_data[logged_in_pin]} BITCOIN")).pack(pady=5)
    tk.Button(services_window, text="Account Details", width=20, command=lambda: messagebox.showinfo("Account Details", "Account Name: John Doe\nAccount Number: 1234567890")).pack(pady=5)
    tk.Button(services_window, text="Back to Main Menu", width=20, command=lambda: [services_window.destroy(), show_main_menu()]).pack(pady=10)

    services_window.mainloop()


# Function to handle transfers
def transfer(parent_window):
    parent_window.destroy()

    transfer_window = tk.Tk()
    transfer_window.title("Transfer")

    tk.Label(transfer_window, text="Transfer Funds", font=("Helvetica", 16, "bold")).pack(pady=10)

    tk.Label(transfer_window, text="Enter amount:").pack()
    amount_entry = tk.Entry(transfer_window)
    amount_entry.pack()

    tk.Label(transfer_window, text="Enter account number:").pack()
    account_entry = tk.Entry(transfer_window)
    account_entry.pack()

    tk.Label(transfer_window, text="Enter bank name:").pack()
    bank_entry = tk.Entry(transfer_window)
    bank_entry.pack()

    tk.Label(transfer_window, text="Enter recipient name:").pack()
    name_entry = tk.Entry(transfer_window)
    name_entry.pack()

    def process_transfer():
        try:
            amount = int(amount_entry.get())
            if amount > user_data[logged_in_pin]:
                messagebox.showerror("Insufficient Funds", "Transfer cannot be processed.")
                return

            account_no = account_entry.get()
            if not account_no.isdigit():
                messagebox.showerror("Invalid Input", "Account number must be numeric.")
                return

            bank = bank_entry.get()
            if bank.isdigit():
                messagebox.showerror("Invalid Input", "Bank name cannot be numeric.")
                return

            name = name_entry.get()
            if name.isdigit():
                messagebox.showerror("Invalid Input", "Recipient name cannot be numeric.")
                return

            # Deduct amount and confirm transfer
            user_data[logged_in_pin] -= amount
            messagebox.showinfo("Transfer Successful", f"Transfer of {amount} BITCOIN to {name} at {bank} is successful.\nNew balance: {user_data[logged_in_pin]} BITCOIN")
            transfer_window.destroy()
            show_main_menu()

        except ValueError:
            messagebox.showerror("Invalid Input", "Please enter valid numeric values for the amount.")

    tk.Button(transfer_window, text="Submit", command=process_transfer).pack(pady=10)
    tk.Button(transfer_window, text="Cancel", command=lambda: [transfer_window.destroy(), show_main_menu()]).pack(pady=5)

    transfer_window.mainloop()


# Initialize login window
login_window = tk.Tk()
login_window.title("Login")

tk.Label(login_window, text="Welcome to Spark Financial Organization", font=("Helvetica", 16, "bold")).pack(pady=10)
tk.Label(login_window, text="Enter your PIN to login:").pack()

pin_entry = tk.Entry(login_window, show="*")
pin_entry.pack()

attempts = tk.IntVar(value=5)  # Remaining login attempts

tk.Button(login_window, text="Login", command=login).pack(pady=10)
tk.Button(login_window, text="Quit", command=login_window.destroy).pack()

login_window.mainloop()
```
